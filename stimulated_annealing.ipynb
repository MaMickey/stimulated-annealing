```python
import math
import random
```

### 2a 
### Find the largest value of z and the corresponding u, Temperature

- The task begin with generating a start point of the user-erntered u value.
- To generate a range of 100 points around the start to search for a larger point than the start point.
- Iterate the second operation until the cooling of the temperature is out of the temperature limitation.
- The final finding of the "start" is the largest value with the corresponding value. 

```python
def stimulated_annealing1(u,T=10000,cool= 0.99):
        # set the start point with the inputed u
        start = u * math.sin(1/(0.01+ math.pow(u,2))) + math.pow(u,3) * math.sin(1/(0.01+ math.pow(u,4)))
        while T > 0.05:
            #generate a range of 100 points to make comparsion to the start point and find the lager value
            for i in range(100):
                #random new u values
                direction_u = random.uniform(-1,1)
                #generate the new points with the new u values
                s1 = direction_u * math.sin(1/(0.01+ math.pow(direction_u,2))) + math.pow(direction_u,3) * math.sin(1/(0.01+ math.pow(direction_u,4)))
                if s1 > start:
                    start = s1
                if s1 == start:
                    start = s1
                #if the random number is still smaller than the acceptable values (the increment of cunrrent and old points divided by Temperature)
                # still accept the new points
                elif random.random() < math.exp((s1 - start) / T):
                    start = s1
            #cooling down the Temperature
            T*=cool
            print("z : %s" %start, "u : %s" %direction_u, "Temperature : %s" %T)      
```

### 2b
### Find the largest value of z and the corresponding u,v,Temperature

- The task begin with generating a start point of the user-erntered u and v value.
- To generate a range of 100 points around the start to search for a larger point than the start point.
- Iterate the second operation until the cooling of the temperature is out of the temperature limitation.
- The final finding of the "start" is the largest value with the corresponding value. 

```python
def stimulated_annealing2(u,v,T=10000,cool= 0.99):
        # set the start point with the inputed u and v
        start = u*math.pow(v,2)*math.sin(v/(0.01+ math.pow(u,2))) + math.pow(u,3)*math.pow(v,2)*math.sin(math.pow(v,3)/(0.01+ math.pow(u,4)))
        while T > 0.05:
            #generate a range of 100 points to make comparsion to the start point and find the lager value
            for i in range(100):
                #random new u values
                direction_u = random.uniform(-1,1)
                #random new v values
                direction_v = random.uniform(-1,1)
                #generate the new points with the new u and v values
                s1 = direction_u*math.pow(direction_v,2)*math.sin(direction_v/(0.01+ math.pow(direction_u,2))) + math.pow(direction_u,3)*math.pow(direction_v,2)*math.sin(math.pow(direction_v,3)/(0.01+ math.pow(direction_u,4)))
                if s1 > start:
                    start = s1
                if s1 == start:
                    start = s1
                #if the random number is still smaller than the acceptable values (the increment of cunrrent and old points divided by Temperature)
                # still accept the new points
                elif random.random() < math.exp((s1 - start) / T):
                    start = s1
            #cooling down the Temperature
            T*=cool
            print("z : %s" %start, "u : %s" %direction_u, "v : %s" %direction_v, "Temperature : %s" %T)
```

### 2c
### Find the largest value of z and the corresponding u,v,w,Temperature

- The task begin with generating a start point of the user-erntered u,v and w value.
- To generate a range of 100 points around the start to search for a larger point than the start point.
- Iterate the second operation until the cooling of the temperature is out of the temperature limitation.
- The final finding of the "start" is the largest value with the corresponding value. 

```python
def stimulated_annealing3(u,v,w,T=10000,cool= 0.99):
        # set the start point with the inputed u,v and w
        start = (u*math.pow(v,2) + math.sin(math.pi*w))*math.sin(v/(0.01+ math.pow(u,2)))*math.sin(math.pi*w/2) \
                + math.pow(u,3)*math.pow(v,2)*w*math.sin(math.pow(v,3)/(0.01*math.pow(math.sin(math.pi*w/2),2) \
                + math.pow(u,4) + math.pow(w-1,2)))
        while T > 0.05:
            #generate a range of 100 points to make comparsion to the start point and find the lager value
            for i in range(100):
                #random new u values
                dir_u = random.uniform(-1,1)
                #random new v values
                dir_v = random.uniform(-1,1)
                #random new w values
                dir_w = random.uniform(-1,1) 
                #generate the new points with the new u,v and w values
                s1 = (dir_u*math.pow(dir_v,2) + math.sin(math.pi*dir_w))*math.sin(dir_v/(0.01+ math.pow(dir_u,2)))*math.sin(math.pi*dir_w/2) \
                + math.pow(dir_u,3)*math.pow(dir_v,2)*dir_w*math.sin(math.pow(dir_v,3)/(0.01*math.pow(math.sin(math.pi*dir_w/2),2) \
                + math.pow(dir_u,4) + math.pow(dir_w-1,2)))
                if s1 > start:
                    start = s1
                if s1 == start:
                    start = s1
                #if the random number is still smaller than the acceptable values (the increment of cunrrent and old points divided by Temperature)
                # still accept the new points
                elif random.random() < math.exp((s1 - start) / T):
                    start = s1
            #cooling down the Temperature
            T*=cool
            print("z : %s" %start, "u : %s" %dir_u, "v : %s" %dir_v, "w : %s" %dir_w, "Temperature : %s" %T)
```

### 2d
### Find the largest value of z and the corresponding u,v,w,Temperature

- The task begin with generating a start point of the user-erntered u,v,w and y value.
- To generate a range of 100 points around the start to search for a larger point than the start point.
- Iterate the second operation until the cooling of the temperature is out of the temperature limitation.
- The final finding of the "start" is the largest value with the corresponding value. 

```python
def stimulated_annealing4(u,v,w,y,T=10000,cool= 0.99):
        # set the start point with the inputed u,v,w and y
        start = ((u*math.pow(v,2) + math.sin(math.pi*w))*math.sin(v/(0.01+ math.pow(u,2)))*math.sin(math.pi*w/2) \
                + math.pow(u,3)*math.pow(v,2)*w*math.sin(math.pow(v,3)/(0.01*math.pow(math.sin(math.pi*w/2),2) \
                + math.pow(u,4) + math.pow(w-1,2))))*y
        while T > 0.05:
            #generate a range of 100 points to make comparsion to the start point and find the lager value
            for i in range(100):
                #random new u values
                dir_u = random.uniform(-1,1)
                #random new v values
                dir_v = random.uniform(-1,1)
                #random new w values
                dir_w = random.uniform(-1,1)
                #random new y values
                dir_y = random.choice([-1,0,1])
                #generate the new points with the new u,v,w and y values
                s1 = ((dir_u*math.pow(dir_v,2) + math.sin(math.pi*dir_w))*math.sin(dir_v/(0.01+ math.pow(dir_u,2)))*math.sin(math.pi*dir_w/2) \
                + math.pow(dir_u,3)*math.pow(dir_v,2)*dir_w*math.sin(math.pow(dir_v,3)/(0.01*math.pow(math.sin(math.pi*dir_w/2),2) \
                + math.pow(dir_u,4) + math.pow(dir_w-1,2))))*dir_y
                if s1 > start:
                    start = s1
                if s1 == start:
                    start = s1
                #if the random number is still smaller than the acceptable values (the increment of cunrrent and old points divided by Temperature)
                # still accept the new points
                elif random.random() < math.exp((s1 - start) / T):
                    start = s1
            #cooling down the Temperature
            T*=cool
            print("z : %s" %start, "u : %s" %dir_u, \
                  "v : %s" %dir_v, "w : %s" %dir_w, \
                  "y : %s" %dir_y, "Temperature : %s" %T)
```
